# Ordinal Encoding

```python
import pandas as pd
```

```python
df = pd.DataFrame({"Size":["s","m","l","xl","s","m","l","s","s","l","xl","m"]}) 
df                                                                # size se columns bana rahe h
```

```python
df.head(3)
```

```python
#encoding karne ke leye order decide kar na hoga hame
ord_data = [["s","m","l","xl"]]   # yaha par oder diside karte kis order me karne nominal me direct tha
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=ord_data)
oe.fit(df[["Size"]])
```

```python
oe.transform(df[["Size"]])  # encode hogaya yaha
```

```python
df["Size_en"] = oe.transform(df[["Size"]])
df
```

# Outlier Removal
- IQR METHOR USE

```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\Users\AUSU\Downloads\loan_data_set.csv")
dataset.head(5)  
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(x="CoapplicantIncome",data=dataset)
plt.show()
```

```python
dataset.shape
```

```python
q1 = dataset["CoapplicantIncome"].quantile(0.25)
q3 = dataset["CoapplicantIncome"].quantile(0.75)
```

```python
IQR = q3 - q1
```

```python
min_range = q1 - (1.5*IQR)
max_range = q3 + (1.5*IQR)
min_range , max_range
```

```python
new_dataset=dataset[dataset["CoapplicantIncome"]<=max_range]
new_dataset
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="CoapplicantIncome",data=new_dataset)
plt.show()
```

```python
dataset.describe()
```

```python
sns.boxplot(x = "CoapplicantIncome",data=dataset)
```

```python
sns.distplot(dataset["CoapplicantIncome"])         #see tail and there is outlier                    # distribution curve banata h (normal distribution curve
```

```python
min_range = dataset["CoapplicantIncome"].mean() - (3*dataset["CoapplicantIncome"].std())
max_range = dataset["CoapplicantIncome"].mean() + (3*dataset["CoapplicantIncome"].std())
```

```python
min_range,max_range    # negative no me outlier nhi present rahta h
```

```python
dataset[dataset["CoapplicantIncome"]<=max_range]  # outlier data hat jaye ga
```
